```python
simulations = ["Single neuron with low noise", "Single neuron with high noise", "Two neurons with the same parameters", "Two neurons with different parameters", "Two synchronsied neurons"]

data = np.array([[126,126,126],[126,92,82],[234,235,230],[165,165,165],[256,244,226]])
actual_number_of_spikes = [256, 165, 234, 126, 126]
detected_spikes = [244, 165, 235, 92, 126]
true_positive_spikes = [226,165, 230, 82, 126]

results = pd.DataFrame(data, index =simulations, columns= ["actual number of spikes", "detected spikes", "true positive spikes"])
results["true positive detection rate"] = results["true positive spikes"]/results["detected spikes"]
results["discriminatory score"] = abs(results["detected spikes"]-results["actual number of spikes"])/ results["actual number of spikes"]
x_pos = [i for i, _ in enumerate(simulations)]

plt.bar(simulations, results["discriminatory score"], color='green')
for index,data in enumerate(results["discriminatory score"]):
    plt.text(x=index -0.26 , y =data +0.01 , s=f"{str(data)[0:5]}" , fontdict=dict(fontsize=15))
plt.xticks(rotation=20, fontsize = 12)
plt.title("Wave_clus positive detection rate for different synthetic data scenarios",  fontsize = 12)
plt.ylim([0.6,None])
plt.show()

```
